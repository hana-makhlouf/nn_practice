## Implementation of a NN:

* feedforward: set of input features and RANDOM weights. This is to test so that the algo can give us errors in backprop
* backpropagation: error between predicted and target so we can correct it further

#### Goal is to minimise error by either increasing or decreasing the weight

## Sigmoid function

#### Used to get an output value of 0 or 1. For any input value, the output will be between 0 and 1

## Methods of calculating error

* mean squared error: sum(0.5*(target-output)^2)

## BIAS (and why we add it)

#### 0,0 will output 0,0 regardless of how many times we train the model. Adding bias allows us to see the results better and make reliable predictions
